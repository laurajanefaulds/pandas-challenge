# PyCity Schools Analysis

Analyzing the average grades, passing rates, budget, sizes and types of the fifteen high schools comprising the PyCity school district has given us the following results:

On average, schools throughout the district all show slightly higher scores in reading (81%) than in math (78%), with a passing rate of 74% in math and 81% in reading, indicating that there is a greater range in unique math scores among students. The average overall passing rate for both subjects is 65%, showing us that there is some disparity between student success rates in either subject, and that high reading scores do not necessarily equal high math scores, and vice versa. 

The PyCity school district is composed of two types of schools, district and charter. Examining the data further, it becomes clear that PyCity’s charter schools are experiencing significantly greater success in overall passing rates than the district schools, with at least 89% of all charter school students passing both math and reading. District schools, meanwhile, are looking at a 53-54% overall pass rate across all schools. 

Individual math and reading scores in district schools are also significantly lower than they are in charter schools.

Across both district and charter schools, math and reading scores remain consistent across all grades, with no significant peaks or valleys. 

Delving deeper into our financial data, we see that the majority of district schools have significantly higher budgets than do the charter schools in PyCity, though they also have uniformly larger numbers of total students. Though our district school students do ultimately receive slightly higher per capita spending than do our charter school students, this does not appear to help their overall success rates. The students in both districts who receive the least funding per capita  experience substantially higher pass rates (90 percent on average) than do the students who receive the most funding per capita, which results in a 53% pass rate. 

A more significant variable when assessing overall passing rates is the school size. Schools with student body populations of less than 1000 experience an average overall passing rate of 90\%, whereas schools with student body populations of 2000-5000 experience an average overall passing rate of 58\%. All district schools in PyCity fall into the 2000-5000 range, while only one charter school falls into this range, with the majority comprised of 2000 students or less. 

Overall, it appears that student body size has a greater bearing on student success rates than does the per capita budget. Our analysis makes clear that, when it comes to student success, money simply does not go as far as one-on-one attention.  


```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load 
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete = school_data_complete.rename(columns={"school_name": "School Name"})
school_data_complete
```

```python
# Calculate the total number of unique schools
school_count = len(school_data_complete["School Name"].unique())
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete["student_name"].count()
student_count
```

```python
# Calculate the total budget
per_school_budget = school_data_complete.groupby("School Name").mean(numeric_only=True)["budget"]
total_budget = per_school_budget.sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score 
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)

MIN_PASSING_SCORE = 70
passing_math_series = school_data_complete["math_score"] >= MIN_PASSING_SCORE
passing_math_count = school_data_complete.loc[passing_math_series, "student_name"].count()
passing_math_percentage = 100 * passing_math_count / student_count
passing_math_percentage

```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  

MIN_READING_SCORE = 70
passing_reading_series = school_data_complete["reading_score"] >= MIN_PASSING_SCORE
passing_reading_count = school_data_complete.loc[passing_reading_series, "student_name"].count()
passing_reading_percentage = 100 * passing_reading_count / student_count
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({"Total Schools": [school_count],
                                 "Total Students": student_count,
                                "Total Budget": total_budget,
                                "Average Math Score": average_math_score,
                                "Average Reading Score": average_reading_score,
                                "% Passing Math": passing_math_percentage,
                                "% Passing Reading": passing_reading_percentage,
                                "% Overall Passing": overall_passing_rate})
# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
school_spending_df
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
#print(spending_math_scores)
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
#print(spending_reading_scores)
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
#print(spending_passing_math)
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
#print(spending_passing_reading)
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()
#print(overall_passing_spending)
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame ({"Math Scores": spending_math_scores,
                                     "Reading Scores": spending_reading_scores,
                                     "% Passing Math": spending_passing_math,
                                     "% Passing Reading": spending_passing_reading,
                                 "% Overall Passing": overall_passing_spending })

# Display results
spending_summary
```
